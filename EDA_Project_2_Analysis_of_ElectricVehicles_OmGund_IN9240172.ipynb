There are fewer extreme values (very high or very low ranges) compared to a normal distribution.

# **3. Bivariate Analysis**

```python
#Is there a relationship between the electric range of an EV and its base MSRP?

import matplotlib.pyplot as plt
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Electric Range', y='Base MSRP', data=ev)
plt.title('Relationship between Electric Range and Base MSRP')
plt.xlabel('Electric Range')
plt.ylabel('Base MSRP')
plt.show()

correlation = ev['Electric Range'].corr(ev['Base MSRP'])
print(f"Correlation between Electric Range and Base MSRP: {correlation}")
```

**Observation notes:**

The correlation coefficient of 0.085 indicates a very weak positive correlation between Electric Range and Base MSRP. This means that there is a slight tendency for EVs with higher electric ranges to have slightly higher base MSRPs. However, the relationship is not strong, and there is a lot of variability in the data.


```python
#Does the make of an EV affect its electric range?


import matplotlib.pyplot as plt
import scipy.stats as stats

# Box plot to compare electric range across different makes
plt.figure(figsize=(12, 6))
sns.boxplot(x='Make', y='Electric Range', data=ev)
plt.title('Electric Range Distribution by Make')
plt.xlabel('Make')
plt.ylabel('Electric Range')
plt.xticks(rotation=90)
plt.show()

# ANOVA test to determine if there are significant differences in electric range among makes
groups = [ev['Electric Range'][ev['Make'] == make] for make in ev['Make'].unique()]
f_statistic, p_value = stats.f_oneway(*groups)

print(f"F-statistic: {f_statistic}")
print(f"P-value: {p_value}")

# Interpretation of results
if p_value < 0.05:
    print("There is a statistically significant difference in electric range among different makes.")
else:
    print("There is no statistically significant difference in electric range among different makes.")
```

```python
#How are EVs distributed across different counties or cities?


import matplotlib.pyplot as plt
# Frequency table of EVs by county
county_counts = ev['County'].value_counts()
print("EVs by County:\n", county_counts)

# Bar chart of EVs by county
plt.figure(figsize=(12, 6))
sns.countplot(x='County', data=ev)
plt.title('Number of EVs by County')
plt.xlabel('County')
plt.ylabel('Number of EVs')
plt.xticks(rotation=90)
plt.show()


# Frequency table of EVs by city
city_counts = ev['City'].value_counts()
print("\nEVs by City:\n", city_counts)

# Bar chart of EVs by city (limiting to top 20 for better visualization)
top_20_cities = city_counts.head(20)
plt.figure(figsize=(12, 6))
sns.countplot(x='City', data=ev[ev['City'].isin(top_20_cities.index)])
plt.title('Number of EVs by City (Top 20)')
plt.xlabel('City')
plt.ylabel('Number of EVs')
plt.xticks(rotation=90)
plt.show()


# Percentage of each EV type in different counties
county_ev_type_percentage = ev.groupby(['County', 'Electric Vehicle Type'])['Electric Vehicle Type'].count() \
                                / ev.groupby('County')['Electric Vehicle Type'].count() * 100
print("\nPercentage of EV Type by County:\n", county_ev_type_percentage)


# Percentage of each EV type in different cities
city_ev_type_percentage = ev.groupby(['City', 'Electric Vehicle Type'])['Electric Vehicle Type'].count() \
                              / ev.groupby('City')['Electric Vehicle Type'].count() * 100
print("\nPercentage of EV Type by City:\n", city_ev_type_percentage)
```

```python
#Has the electric range of EVs increased over the years?


import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats

# Scatter plot to visualize the trend
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Model Year', y='Electric Range', data=ev)
plt.title('Electric Range of EVs Over the Years')
plt.xlabel('Model Year')
plt.ylabel('Electric Range')
plt.show()

# Correlation coefficient to measure the relationship
correlation = ev['Model Year'].corr(ev['Electric Range'])
print(f"Correlation between Model Year and Electric Range: {correlation}")

# Regression analysis to model the trend
slope, intercept, r_value, p_value, std_err = stats.linregress(ev['Model Year'], ev['Electric Range'])
print(f"Slope: {slope}")
print(f"Intercept: {intercept}")
print(f"R-squared: {r_value**2}")
print(f"P-value: {p_value}")

# Plot the regression line
plt.figure(figsize=(8, 6))
sns.regplot(x='Model Year', y='Electric Range', data=ev)
plt.title('Electric Range of EVs Over the Years with Regression Line')
plt.xlabel('Model Year')
plt.ylabel('Electric Range')
plt.show()

# Interpretation
if correlation > 0 and p_value < 0.05:
  print("There is a statistically significant positive correlation between model year and electric range, suggesting that electric range has increased over the years.")
else:
  print("There is no statistically significant correlation between model year and electric range.")
```

```python
#How does the make of an EV affect the base MSRP?


import matplotlib.pyplot as plt
# Box plot to compare base MSRP across different makes
plt.figure(figsize=(12, 6))
sns.boxplot(x='Make', y='Base MSRP', data=ev)
plt.title('Base MSRP Distribution by Make')
plt.xlabel('Make')
plt.ylabel('Base MSRP')
plt.xticks(rotation=90)
plt.show()

# ANOVA test to determine if there are significant differences in Base MSRP among makes
groups = [ev['Base MSRP'][ev['Make'] == make] for make in ev['Make'].unique()]
f_statistic, p_value = stats.f_oneway(*groups)

print(f"F-statistic: {f_statistic}")
print(f"P-value: {p_value}")

# Interpretation of results
if p_value < 0.05:
    print("There is a statistically significant difference in Base MSRP among different makes.")
else:
    print("There is no statistically significant difference in Base MSRP among different makes.")
```

```python
"""Is there an association between the type of electric vehicle
and its eligibility for Clean Alternative Fuel Vehicle (CAFV) incentives or
programs?"""

import pandas as pd
import matplotlib.pyplot as plt
# Create a contingency table to examine the relationship between EV type and CAFV eligibility
contingency_table = pd.crosstab(ev['Electric Vehicle Type'], ev['Clean Alternative Fuel Vehicle (CAFV) Eligibility'])

# Display the contingency table
print(contingency_table)

# Perform a Chi-squared test of independence to determine if there's an association
chi2, p_value, _, _ = stats.chi2_contingency(contingency_table)

print(f"\nChi-squared statistic: {chi2}")
print(f"P-value: {p_value}")

# Interpretation of the results
if p_value < 0.05:
    print("\nThere is a statistically significant association between the type of electric vehicle and its CAFV eligibility.")
else:
    print("\nThere is no statistically significant association between the type of electric vehicle and its CAFV eligibility.")

# Visualize the association using a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(contingency_table, annot=True, fmt='d', cmap='viridis')
plt.title('Association between EV Type and CAFV Eligibility')
plt.xlabel('CAFV Eligibility')
plt.ylabel('Electric Vehicle Type')
plt.show()
```

# **4. Create a Choropleth using plotly.express to display the number of EV vehicles based on location**

```python
import plotly.express as px

# Assuming 'County' column represents the location
ev_county_counts = ev.groupby('County')['VIN'].count().reset_index()

# Create the Choropleth map
fig = px.choropleth(ev_county_counts,
                    locations='County',  # Column with location names
                    locationmode='USA-states',  # Specify location mode for US states
                    color='VIN',  # Column with values to be represented by color
                    hover_name='County',  # Column for hover text
                    color_continuous_scale='Viridis',  # Choose a color scale
                    title='Number of Electric Vehicles by County')

fig.update_layout(
    geo_scope='usa',  # Set the scope to the USA
)

fig.show()
```
